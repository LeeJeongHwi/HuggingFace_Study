```python
def convert_examples_to_features(examples):

    # dataset의 각 레코드마다 en, ko 특성에 추가로 input_ids, attention_mask, label 특성이 추가됨
    # truncation : Input Sequence가 너무 길어지는것을 방지, 모델이 처리할 수 있는 고정된 길이를 유지하도록 사용
    #               역으로 Sequence가 너무 길면 Padding으로 길이를 맞추기도함

    model_inputs = tokenizer(examples["en"],
                             text_target=examples["ko"],
                             max_length=max_token_length, truncation=True)
    return model_inputs
```

```python
# Thread 개수 지정
NUM_CPU = multiprocessing.cpu_count()
NUM_CPU
```

```python
tokenized_datasets = dataset.map(convert_examples_to_features,
                                 batched=True,
                                 remove_columns=dataset["train"].column_names,
                                 num_proc=NUM_CPU)
# Remove Column을 해주지않는다면 Collator가 Sample들을 Mini batch로 묶을 때 패딩 처리 X  --> 계속 "en", "ko"가 남기 때문
tokenized_datasets
```

## Model

```python
model = AutoModelForSeq2SeqLM.from_pretrained(model_ckpt).to(device)

# in_features = 768, out_features = 64128 (CLS HEAD)
```

```python
model
```

```python
# Transformer를 돌리기위한 테스트는 Encoder, Decoder에 Forwarding Test부터 해야한다. (직접 만든 모델이 아니기 때문임)
# 어떤 구조인지 파악하고 input shape을 맞추기 위함

# Forwading test
encoder_inputs = tokenizer(["Studies have been shown that owning a dog is good for you"], return_tensors="pt")["input_ids"].to(device)
decoder_targets = tokenizer(["개를 키우눈 것이 건강에 좋다는 연구 결과가 있습니다."], return_tensors="pt")["input_ids"].to(device)

print(encoder_inputs)
print(decoder_targets)
```

```python
decoder_inputs = model._shift_right(decoder_targets) # Teacher Forcing

# Teacher Forcing이란?
"""
원래 Decoder는 현재 시점의 출력을 생성하기 위해 이전 시점의 출력을 입력으로 받음 -> "이전 시점의 출력"을 "이전 시점의 실제 정답값"으로 대체
즉, 각 타임 스텝에서 모델이 예측한 값이 아닌 "실제 정답값"으로 입력을 사용하는 것
'Seq2Seq의 핵심적인 기능'
"""
```

```python
pd.DataFrame([tokenizer.convert_ids_to_tokens(decoder_targets[0]),
              tokenizer.convert_ids_to_tokens(decoder_inputs[0])], # added Padding
             index=("decoder target", "decoder input")
)
```

```python
outputs = model(input_ids=encoder_inputs, decoder_input_ids=decoder_inputs, labels=decoder_targets) # Forwarding Pass
```

```python
print(outputs["loss"])
print(outputs["logits"].shape) # Token Prediction Values, 1 sample 10 Token 64128 Words
print(outputs["encoder_last_hidden_state"].shape) # Encoder Last out Shape
```

```python
ans = tokenizer.convert_ids_to_tokens(torch.argmax(outputs["logits"][0], axis=1).cpu().numpy())
print(ans, len(ans))
# 아직 CLS HEADER가 학습이 안되어서 이상한 값이 나오는 것
```

### Collator

모델 학습시키기 위해서는 `DataLoader`를 사용하는데, Dataset으로부터 Sample을 Minibatch 만큼 가져오기 위해 사용한다. 배치수 만큼 부작위로 가져와서 샘플을 짝맞춤해서 반환해야함
서로 다른 샘플 크기라면, 반환전 크기, 길이, 패딩추가 등 작업이 필요함

## Trainer

```python
training_args = Seq2SeqTrainingArguments(
    output_dir="chkpt", # chkpt 저장 위치
    learning_rate = 0.0005, # learning rate
    weight_decay=0.01, # weight Decay
    per_device_train_batch_size=32, # Train Batch Size
    per_device_eval_batch_size=32, # Eval Btach Size
    num_train_epochs=1, # train epoch??
    save_steps=500, # Checkpoint지점
    save_total_limit=2, # Checkpoint 최대 개수
    evaluation_strategy="epoch",# evaluation 전략?
    logging_strategy="no", # logging 전략
    predict_with_generate=True, # 평가 시 생성된 텍스트를 사용하여 예측
    fp16=False, # 16비트 부동소수점 사용여부
    gradient_accumulation_steps=2, # 큰 배치 크기를 시뮬레이션 하기 위해 사용, Gradient를 누적하고 한번에 가중치 업데이트하는 방식 , GPU 메모리 제한에 사용
    report_to="none" # wandb log off
)
```

```python
def compute_metrics(eval_preds):
    preds, labels = eval_preds

    if isinstance(preds, tuple):
        preds = preds[0]

    decoded_preds = tokenizer.batch_decode(preds, skip_special_tokens=True)

    # Replace -100 in the labels as we can't decode them.
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    # Some simple post-processing
    decoded_preds = [pred.strip() for pred in decoded_preds]
    decoded_labels = [[label.strip()] for label in decoded_labels]

    result = metric.compute(predictions=decoded_preds, references=decoded_labels)
    result = {"bleu": result["score"]}

    return result
```

```python
trainer = Seq2SeqTrainer(
    model,
    training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["valid"],
    data_collator=data_collator,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```
