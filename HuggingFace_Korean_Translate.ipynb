```python
def convert_examples_to_features(examples):

    # dataset의 각 레코드마다 en, ko 특성에 추가로 input_ids, attention_mask, label 특성이 추가됨
    # truncation : Input Sequence가 너무 길어지는것을 방지, 모델이 처리할 수 있는 고정된 길이를 유지하도록 사용
    #               역으로 Sequence가 너무 길면 Padding으로 길이를 맞추기도함

    model_inputs = tokenizer(examples["en"],
                             text_target=examples["ko"],
                             max_length=max_token_length, truncation=True)
    return model_inputs
```

```python
# Thread 개수 지정
NUM_CPU = multiprocessing.cpu_count()
NUM_CPU
```

```python
tokenized_datasets = dataset.map(convert_examples_to_features,
                                 batched=True,
                                 remove_columns=dataset["train"].column_names,
                                 num_proc=NUM_CPU)
# Remove Column을 해주지않는다면 Collator가 Sample들을 Mini batch로 묶을 때 패딩 처리 X  --> 계속 "en", "ko"가 남기 때문
tokenized_datasets
```

## Model

```python
model = AutoModelForSeq2SeqLM.from_pretrained(model_ckpt).to(device)

# in_features = 768, out_features = 64128 (CLS HEAD)
```

```python
model
```

```python
# Transformer를 돌리기위한 테스트는 Encoder, Decoder에 Forwarding Test부터 해야한다. (직접 만든 모델이 아니기 때문임)
# 어떤 구조인지 파악하고 input shape을 맞추기 위함

# Forwading test
encoder_inputs = tokenizer(["Studies have been shown that owning a dog is good for you"], return_tensors="pt")["input_ids"].to(device)
decoder_targets = tokenizer(["개를 키우눈 것이 건강에 좋다는 연구 결과가 있습니다."], return_tensors="pt")["input_ids"].to(device)

print(encoder_inputs)
print(decoder_targets)
```

```python
decoder_inputs = model._shift_right(decoder_targets) # Teacher Forcing

# Teacher Forcing이란?
"""
원래 Decoder는 현재 시점의 출력을 생성하기 위해 이전 시점의 출력을 입력으로 받음 -> "이전 시점의 출력"을 "이전 시점의 실제 정답값"으로 대체
즉, 각 타임 스텝에서 모델이 예측한 값이 아닌 "실제 정답값"으로 입력을 사용하는 것
'Seq2Seq의 핵심적인 기능'
"""
```

```python
pd.DataFrame([tokenizer.convert_ids_to_tokens(decoder_targets[0]),
              tokenizer.convert_ids_to_tokens(decoder_inputs[0])], # added Padding
             index=("decoder target", "decoder input")
)
```

```python
outputs = model(input_ids=encoder_inputs, decoder_input_ids=decoder_inputs, labels=decoder_targets) # Forwarding Pass
```

```python
print(outputs["loss"])
print(outputs["logits"].shape) # Token Prediction Values, 1 sample 10 Token 64128 Words
print(outputs["encoder_last_hidden_state"].shape) # Encoder Last out Shape
```
